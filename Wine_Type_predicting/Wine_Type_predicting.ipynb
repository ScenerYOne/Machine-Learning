### Outliers Handling


```python
# การจัดการ Outliers ด้วย IQR
def remove_outliers_iqr(df, column):
    # หาค่า Q1 (ควอไทล์ที่ 25) และ Q3 (ควอไทล์ที่ 75)
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1  # Interquartile Range

    # กำหนดขอบเขตต่ำสุดและสูงสุดที่ยอมรับได้
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # ลบค่าผิดปกติออก
    df_filtered = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df_filtered
```

```python
# ตรวจสอบฟีเจอร์ที่อาจมีค่าผิดปกติ
columns_to_check = ['Malic acid', 'Alcalinity of ashe', 'Total phenols', 'Color intensity']

# ทำการลบค่าผิดปกติในแต่ละฟีเจอร์
for column in columns_to_check:
    wine_data = remove_outliers_iqr(wine_data, column)

# ตรวจสอบข้อมูลหลังการลบ Outliers
print("ข้อมูลหลังจากการจัดการกับค่าผิดปกติ:")
print(wine_data.describe())
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# เลือกฟีเจอร์ที่สำคัญเพื่อใช้ในการจัดกลุ่ม
selected_features = ['Malic acid', 'Total phenols', 'Flavanoidse', 'Color intensity', 'pH']

# สร้าง DataFrame ที่ใช้ฟีเจอร์เหล่านี้
X = wine_data[selected_features]

# ทำการ Normalize ข้อมูลฟีเจอร์
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ใช้ K-Means Clustering เพื่อจัดกลุ่มไวน์ออกเป็น 2 กลุ่ม (คุณภาพสูงและคุณภาพต่ำ)
kmeans = KMeans(n_clusters=2, random_state=42)
wine_data['quality'] = kmeans.fit_predict(X_scaled)

# แสดงข้อมูลคุณภาพที่สร้างขึ้นใหม่
print(wine_data[['Malic acid', 'Total phenols', 'Flavanoidse', 'Color intensity', 'pH', 'quality']].head())

# ตรวจสอบว่ามีกี่กลุ่มในคุณภาพไวน์
print("\nจำนวนกลุ่มของคุณภาพไวน์:\n", wine_data['quality'].value_counts())
```

### Categorical Data Handling
    1. Label Encoding แปลงค่าหมวดหมู่ในคอลัมน์ WineType เป็นตัวเลข เช่น 'Red' -> 2, 'White' -> 1, 'Rose' -> 0.
    2. One-Hot Encoding แปลงค่าหมวดหมู่ในคอลัมน์ Color เป็นคอลัมน์ตัวเลขหลายคอลัมน์ เช่น 'Dark' -> [1, 0], 'Light' -> [0, 1].


```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# 1. ฟังก์ชันสำหรับสร้างคอลัมน์ใหม่ที่แบ่งประเภทไวน์ตามค่า pH โดยแก้ไขลำดับการตรวจสอบ
def wine_type_by_ph(row):
    if 3.0 <= row['pH'] <= 3.5:  # ไวน์ขาว: ควบคุม pH ให้อยู่ในช่วง 3.0 – 3.5
        return 'White Wine'
    elif 3.51 <= row['pH'] <= 3.8:  # ไวน์แดง: ควบคุม pH ให้อยู่ในช่วง 3.51 – 3.8 (ปรับให้ไม่ทับกับไวน์ขาว)
        return 'Red Wine'
    else:  # หากไม่อยู่ในช่วงที่กำหนด ให้เป็น 'Unknown'
        return 'Unknown'

# เพิ่มคอลัมน์หมวดหมู่ใหม่ที่แสดงประเภทของไวน์
wine_data['Wine_Type'] = wine_data.apply(wine_type_by_ph, axis=1)

# แสดงตัวอย่างข้อมูลและการกระจายตัวของประเภทไวน์
print(wine_data[['pH', 'Wine_Type']].head())
print("\nการกระจายตัวของประเภทไวน์:")
print(wine_data['Wine_Type'].value_counts())

# 2. การแปลงหมวดหมู่ด้วย Label Encoding
label_encoder = LabelEncoder()
wine_data['Wine_Type_Encoded'] = label_encoder.fit_transform(wine_data['Wine_Type'])

# แสดงผลลัพธ์ของการ Label Encoding
print("\nผลลัพธ์หลังจาก Label Encoding:\n", wine_data[['Wine_Type', 'Wine_Type_Encoded']].head())

# 3. การแปลงหมวดหมู่ด้วย One-Hot Encoding
one_hot_encoded = pd.get_dummies(wine_data['Wine_Type'], prefix='Wine')

# รวมผลลัพธ์จาก One-Hot Encoding กลับไปยัง DataFrame หลัก
wine_data = pd.concat([wine_data, one_hot_encoded], axis=1)

# ตรวจหาคอลัมน์ที่ซ้ำและลบออก
wine_data = wine_data.loc[:, ~wine_data.columns.duplicated()]

# แสดงผลลัพธ์ของการ One-Hot Encoding หลังจากลบคอลัมน์ที่ซ้ำ
print("\nผลลัพธ์หลังจาก One-Hot Encoding และลบคอลัมน์ที่ซ้ำ:")
print(wine_data[['Wine_Type'] + list(one_hot_encoded.columns)].head())

# แสดงตัวอย่างข้อมูลที่มีไวน์แดง
print("\nตัวอย่างข้อมูลที่มีไวน์แดง:")
print(wine_data[wine_data['Wine_Type'] == 'Red Wine'][['pH', 'Wine_Type'] + list(one_hot_encoded.columns)].head())

# แสดงตัวอย่างข้อมูลที่เป็น Unknown
print("\nตัวอย่างข้อมูลที่อยู่ในประเภท 'Unknown':")
print(wine_data[wine_data['Wine_Type'] == 'Unknown'][['pH', 'Wine_Type'] + list(one_hot_encoded.columns)].head())
```

### Model Selection
### 1. Logistic Regression

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_auc_score
from sklearn.model_selection import KFold
from tabulate import tabulate
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.preprocessing import label_binarize

# กำหนดจำนวน folds
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# สร้างดิกชันนารีของโมเดลต่าง ๆ ที่ต้องการทดสอบ
classifiers = {
    'Random Forest': RandomForestClassifier(random_state=42),
    'Logistic Regression': LogisticRegression(random_state=42),
    'SVM': SVC(random_state=42, probability=True),  # SVM ใช้ probability=True สำหรับ predict_proba
    'Gradient Boosting': GradientBoostingClassifier(random_state=42)
}

# ตรวจสอบและจัดการกับค่า NaN ก่อนทำการฝึกโมเดล
X_train_cleaned = pd.DataFrame(X_train).fillna(pd.DataFrame(X_train).mean()).values

# ใช้ X_train_cleaned ในการฝึกโมเดล
for name, clf in classifiers.items():
    fold = 0
    print(f'==================== {name} ====================')

    for train_index, test_index in kf.split(X_train_cleaned):
        # แยกข้อมูล train และ test สำหรับ fold นี้
        X_fold_train, X_fold_test = X_train_cleaned[train_index], X_train_cleaned[test_index]
        y_fold_train, y_fold_test = y_train.iloc[train_index], y_train.iloc[test_index]

        # ฝึกโมเดล
        clf.fit(X_fold_train, y_fold_train)

        # ทำนายผลลัพธ์
        y_pred = clf.predict(X_fold_test)

        # ตรวจสอบว่าโมเดลมีฟังก์ชัน predict_proba หรือไม่
        if hasattr(clf, "predict_proba"):
            y_pred_proba = clf.predict_proba(X_fold_test)
        else:
            y_pred_proba = None  # ถ้าไม่มี predict_proba, ไม่คำนวณ ROC-AUC

        # สร้าง Confusion Matrix
        cm = confusion_matrix(y_fold_test, y_pred)

        print(f'Fold {fold + 1}')

        # แสดง Confusion Matrix
        plt.figure(figsize=(6, 4))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
        plt.title(f'{name} - Confusion Matrix (Fold {fold + 1})')
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.show()

        # คำนวณตัวชี้วัด
        accuracy = accuracy_score(y_fold_test, y_pred)
        precision = precision_score(y_fold_test, y_pred, average='weighted')
        recall = recall_score(y_fold_test, y_pred, average='weighted')
        f1 = f1_score(y_fold_test, y_pred, average='weighted')

        # ตรวจสอบการคำนวณ ROC-AUC สำหรับ multiclass classification
        if y_pred_proba is not None:
            # ทำ one-hot encoding สำหรับ y_test
            y_fold_test_binarized = label_binarize(y_fold_test, classes=np.unique(y_fold_test))
            if y_fold_test_binarized.shape[1] > 1:  # เช็คว่ามีหลายคลาส
                roc_auc = roc_auc_score(y_fold_test_binarized, y_pred_proba, multi_class='ovr')  # สำหรับ multiclass classification
            else:
                roc_auc = "N/A"
        else:
            roc_auc = "N/A"

        # จัดเก็บตัวชี้วัดในรูปแบบตาราง
        metrics = [
            ["Accuracy", f"{accuracy:.2f}"],
            ["Precision", f"{precision:.2f}"],
            ["Recall", f"{recall:.2f}"],
            ["F1-score", f"{f1:.2f}"],
            ["ROC-AUC", f"{roc_auc}" if isinstance(roc_auc, str) else f"{roc_auc:.2f}"]
        ]

        # แสดงผลในรูปแบบตาราง
        print(tabulate(metrics, headers=["Metric", "Score"], tablefmt="pretty"))
        print()
        fold += 1
```
